```
def create_list_sentences(text):
  sent = []
  doc = nlp(text)
  for s in doc.sents:
    sent.append(str(s).replace(":"," "))
  return sent
```

```
def get_date_count(sent):
  matches = datefinder.find_dates(text = sent, index = True)
  count = 0
  try:
    for match in matches:
      count = count+1
  except:
    pass
  return count

def return_datetime_sentences(list_sent):
  date_sent = []
  count = 0
  for s in list_sent:
    #if('commenc' in s or 'start' in s):
    count = get_date_count(s)
    if(count>0):
      date_sent.append(s)
  return date_sent
```

```
bert_embedding = BertEmbedding()
```

```
def get_embedding(doc):
    sentences = doc.split('.')          
    embeddings = bert_embedding(sentences)
    
    mean_embedding = np.zeros(768)
    n_tokens = 0
    for sent in embeddings:
        n_tokens += len(sent[1])
        mean_embedding += np.sum(np.array(sent[1]), axis=0)       
    mean_embedding /= n_tokens
   
    return mean_embedding
```

```
def cosine_sim(a, b):
    return (dot(a, b)/(norm(a)*norm(b)))
```

```
cosine_sim(get_embedding(text1), get_embedding(text2))
```

```
cosine_sim(get_embedding(text1), get_embedding(text3))
```

```
threshold = 0.75
commence_kamke_sent = []
commence_score = []

interest_kamke_sent = []
interest_score = []

commence_kamka = text1
interest_kamka = text9

commence_kamka_embed = get_embedding(commence_kamka)
interest_kamka_embed = get_embedding(interest_kamka)

for sent in date_sentences:
  embedd_sent = get_embedding(sent)
  commence_cos_value = cosine_sim(commence_kamka_embed, embedd_sent)
  interest_cos_value = cosine_sim(interest_kamka_embed, embedd_sent)
  if commence_cos_value >= threshold:
    commence_kamke_sent.append(sent.replace(",","and"))
    commence_score.append(commence_cos_value)

  if interest_cos_value >= threshold:
    interest_kamke_sent.append(sent.replace(",","and"))
    interest_score.append(interest_cos_value)
```

```
def first_payment_date(sentence):
  interest_regex = re.compile(r'(first)?\s*(?:specified)?\s*(?:interest)?\s*payment\s*date\(*s*\)*')
  mo = interest_regex.search(sentence.lower())
  if mo:
    first_payment = mo.start()
    count=0
    i=0
    matches = datefinder.find_dates(text = sentence, index = True)
    dictionary={}
    dif=[]
    for date,index in matches:
      dictionary[count]=date
      dif.append(abs(index[0]-first_payment))
      count+=1
    print(dictionary[dif.index(min(dif))])
    print(sentence)
  else:
    pass
```

```
def commencing_wala(sentence,commencing_jaisa):
  matches = datefinder.find_dates(text = sentence, index = True)
  for date,index in matches:
    com_index=sentence.find(commencing_jaisa)
    if index[0]>com_index:
      ans=date
      print(ans)
      break
```

```
def final_func(sent):
  if sent.find("commencing")!=-1:
    commencing_wala(sent,"commencing")
    print(sent)
  elif sent.find("beginning")!=-1:
    commencing_wala(sent,"beginning")
    print(sent)
  elif sent.find("starting")!=-1:
    commencing_wala(sent,"starting")
    print(sent)
  else:
    pass
```
